```python
#importing libs
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#loading dataset
dataframe = pd.read_csv("/content/diabetes.csv")
dataframe.head()
```

```python
#preprocessing
df_features = dataframe.drop('Outcome', 1)
df_label = dataframe['Outcome']
#df_features.replace('?', -99999, inplace=True)

```

```python
#classification
label = []
for lab in df_label:
    if lab == 1:
        label.append([1, 0])  # class 1
    elif lab == 0:
        label.append([0, 1])  # class 0

```

```python
data = np.array(df_features)
label = np.array(label)

for a in label:
  print(a)

```

```python
#spliting data
x_train, x_test, y_train, y_test = train_test_split(data, label, test_size=0.2, random_state=42)
x_train.shape

```

```python
#setting parameters
import keras
from matplotlib import pyplot as plt
model = Sequential()
model.add(Dense(500, input_dim=8, activation='sigmoid'))
model.add(Dense(100, activation='sigmoid'))
model.add(Dense(2, activation='softmax'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
history = model.fit(x_train,y_train, epochs=1000, batch_size=70, validation_data=(x_test, y_test))

plt.plot(history.history['accuracy'])
plt.title('эксперимент 4')
plt.ylabel('точность')
plt.xlabel('epoch')
plt.legend(['точность'], loc='upper left')
plt.show()

```

```python
scores = model.evaluate(x_train,y_train)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
